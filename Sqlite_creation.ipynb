```python
from sqlalchemy import create_engine
from sqlalchemy.orm import Session
import pandas
from sqlalchemy.ext.automap import automap_base
```

```python
engine = create_engine("sqlite:///species_complete.sqlite", echo=False)
conn = engine.connect()
```

```python
session = Session(bind=engine)
```

```python
df = pandas.read_csv('Output_data/species_complete.csv')
df.to_sql('species', conn, if_exists='replace')
```

```python
from sqlalchemy import inspect
```

```python
inspector = inspect(engine)
```

```python
inspector.get_table_names()
```

```python
columns = inspector.get_columns('species')
for column in columns:
    print(column["name"], column["type"])
```

```python
engine.execute("SELECT * FROM species WHERE Family='Oculinidae'").fetchall()
```

```python
Base = automap_base()
Base.prepare(engine, reflect=True)
Species = Base.classes.species
```

```python
results = session.query(Species.index).group_by(Species.index)
```

```python
for sp in results:
    print(sp[0])
```
