```python
species = df3['ScientificName'].unique()
print(len(species))
for sp in species:
    print(sp)
```

```python
ids = {}
index=1
total_species = len(species)
for specie in species:
    try:
        url = 'http://www.marinespecies.org/rest/AphiaIDByName/'
        print(f'Getting record {index} of {total_species}')
        print(f'Getting ID for: {specie} ')
        idx = requests.get(url+specie).json()
        print(f'ID: {idx}')
        print('-----------------------------------------')
        ids[specie]=idx
        index = index+1
    except:
        print(f'ID not found')
        print('-----------------------------------------')
        index = index+1
```

```python
total_ids = len(ids)
```

```python
# Function that takes species' ID, scrapes the taxonomy details and makes a list of them
def get_taxa(sp_id):
    try:
        url = 'http://www.marinespecies.org/aphia.php?p=taxdetails&id='
        print(f'Scraping {url} for species ID: {sp_id}')
        response = requests.get(url+str(sp_id))
        soup = bs(response.text, 'html.parser')
        taxa = soup.find_all('ol', class_='breadcrumb')
        taxa_list = taxa[0].find_all('a')
        taxa_comp = []
        for i in taxa_list:
            taxa_comp.append(i.getText())
        print(taxa_comp)
    except:
        print('ID not found')
        pass
    return taxa_comp
    
```

```python
# Iterate through all keys and values (sp name: id) of the ids dict and scrape the taxonomy data using get_taxa function
taxa_dict = {}
cont = 1
for sp, Id in ids.items():
    try:
        print('----------------------------------------')
        print(f'Retrieving record {cont} of {total_ids}')
        taxa_dict[sp] = get_taxa(Id)
        cont += 1
    except:
        cont +=1
        pass
```
